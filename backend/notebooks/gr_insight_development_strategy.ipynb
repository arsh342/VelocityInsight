## ğŸ› Tire Degradation Model Implementation

Now let's implement the tire degradation model - a key component for strategic pit stop recommendations.

### Model Features:
- **Lap Time Progression**: Track performance decline over laps
- **Driving Style Impact**: Aggressive vs conservative degradation patterns  
- **Track-Specific Factors**: Different wear rates per circuit
- **Real-time Predictions**: Estimate remaining tire life

````python
# Tire Degradation Model Implementation
import sys
sys.path.append('../app')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error, r2_score
import warnings
warnings.filterwarnings('ignore')

class TireDegradationModel:
    """
    Model to predict tire performance degradation over race distance.
    Analyzes lap time progression and driving style impact on tire wear.
    """
    
    def __init__(self):
        self.model = None
        self.poly_features = None
        self.degradation_rate = None
        self.baseline_laptime = None
        
    def calculate_lap_degradation(self, df_laps, vehicle_id=None):
        """
        Calculate tire degradation metrics from lap time data.
        
        Args:
            df_laps: DataFrame with lap times and metadata
            vehicle_id: Optional filter for specific vehicle
            
        Returns:
            DataFrame with degradation metrics per lap
        """
        if vehicle_id:
            df_laps = df_laps[df_laps['vehicle_id'] == vehicle_id].copy()
            
        # Sort by timestamp to get lap progression
        df_laps = df_laps.sort_values(['vehicle_id', 'timestamp']).copy()
        
        degradation_data = []
        
        for vehicle, group in df_laps.groupby('vehicle_id'):
            if len(group) < 3:  # Need minimum laps for analysis
                continue
                
            group = group.reset_index(drop=True)
            
            # Calculate lap times (assuming timestamp is lap completion time)
            lap_times = []
            for i in range(1, len(group)):
                lap_time = (group.loc[i, 'timestamp'] - group.loc[i-1, 'timestamp']).total_seconds()
                lap_times.append(lap_time)
            
            if not lap_times:
                continue
                
            # Remove outliers (pit stops, safety cars, etc.)
            lap_times = np.array(lap_times)
            q75, q25 = np.percentile(lap_times, [75, 25])
            iqr = q75 - q25
            lower_bound = q25 - 1.5 * iqr
            upper_bound = q75 + 1.5 * iqr
            
            # Filter outliers
            valid_laps = (lap_times >= lower_bound) & (lap_times <= upper_bound)
            clean_lap_times = lap_times[valid_laps]
            lap_numbers = np.arange(1, len(lap_times) + 1)[valid_laps]
            
            if len(clean_lap_times) < 3:
                continue
                
            # Calculate degradation metrics
            baseline_time = np.min(clean_lap_times)  # Best lap time
            
            for i, (lap_num, lap_time) in enumerate(zip(lap_numbers, clean_lap_times)):
                degradation_data.append({
                    'vehicle_id': vehicle,
                    'lap_number': lap_num,
                    'lap_time': lap_time,
                    'baseline_time': baseline_time,
                    'time_delta': lap_time - baseline_time,
                    'degradation_pct': ((lap_time - baseline_time) / baseline_time) * 100,
                    'cumulative_laps': i + 1
                })
        
        return pd.DataFrame(degradation_data)
    
    def fit_degradation_model(self, degradation_df):
        """
        Fit polynomial regression model to predict tire degradation.
        
        Args:
            degradation_df: DataFrame from calculate_lap_degradation
        """
        if degradation_df.empty:
            raise ValueError("No degradation data available for model fitting")
            
        # Prepare features: lap number, cumulative distance proxy
        X = degradation_df[['lap_number', 'cumulative_laps']].values
        y = degradation_df['degradation_pct'].values
        
        # Use polynomial features to capture non-linear degradation
        self.poly_features = PolynomialFeatures(degree=2, include_bias=False)
        X_poly = self.poly_features.fit_transform(X)
        
        # Fit linear regression on polynomial features
        self.model = LinearRegression()
        self.model.fit(X_poly, y)
        
        # Calculate model performance
        y_pred = self.model.predict(X_poly)
        mae = mean_absolute_error(y, y_pred)
        r2 = r2_score(y, y_pred)
        
        # Calculate average degradation rate (% per lap)
        self.degradation_rate = degradation_df.groupby('lap_number')['degradation_pct'].mean().diff().mean()
        self.baseline_laptime = degradation_df['baseline_time'].mean()
        
        return {
            'mae': mae,
            'r2_score': r2,
            'avg_degradation_rate_per_lap': self.degradation_rate,
            'samples': len(degradation_df)
        }
    
    def predict_degradation(self, lap_number, cumulative_laps):
        """
        Predict tire degradation for given lap.
        
        Args:
            lap_number: Current lap number
            cumulative_laps: Total laps completed
            
        Returns:
            Predicted degradation percentage
        """
        if self.model is None:
            raise ValueError("Model not fitted. Call fit_degradation_model first.")
            
        X = np.array([[lap_number, cumulative_laps]])
        X_poly = self.poly_features.transform(X)
        
        return self.model.predict(X_poly)[0]
    
    def estimate_remaining_performance(self, current_lap, target_laps, baseline_time):
        """
        Estimate performance at target lap based on current degradation.
        
        Args:
            current_lap: Current lap number
            target_laps: Number of additional laps to predict
            baseline_time: Best lap time achieved
            
        Returns:
            Dict with performance predictions
        """
        predictions = []
        
        for i in range(1, target_laps + 1):
            future_lap = current_lap + i
            cumulative = current_lap + i
            
            degradation_pct = self.predict_degradation(future_lap, cumulative)
            predicted_time = baseline_time * (1 + degradation_pct / 100)
            
            predictions.append({
                'lap': future_lap,
                'degradation_pct': degradation_pct,
                'predicted_laptime': predicted_time,
                'time_loss': predicted_time - baseline_time
            })
        
        return predictions

print("ğŸ› Tire Degradation Model implemented!")
print("Features:")
print("  â€¢ Lap time progression analysis")
print("  â€¢ Outlier filtering (pit stops, safety cars)")
print("  â€¢ Polynomial regression for non-linear wear")
print("  â€¢ Performance prediction for future laps")
print("  â€¢ Degradation rate calculation")
````
